# 9- Lowercase all column names

```python
df.columns = df.columns.str.lower()
df.columns
```

# 10- Change all No values to NoRain and all Yes values to Rain in raintoday and raintomorrow

```python
df["raintoday"].replace({"Yes": "Rain", "No": "NoRain"}, inplace=True)
df["raintomorrow"].replace({"Yes": "Rain", "No": "NoRain"}, inplace=True)
df[['raintoday','raintomorrow']]
```

# 11- Change the data type of 'date' (object) column to datetime64 DD/MM/YYYY

```python
df['date'] = df['date'].map(lambda x: datetime.datetime.strptime(x, '%Y-%m-%d').strftime('%d-%m-%Y'))
df['date']
```

# 12- Create a new column called "difference", calculate the difference between maxtemp and mintemp columns for each row, and store the value in this new column.

```python
df['difference'] = df['maxtemp']-df['mintemp']
df['difference'].head()
```

# 13- Remove the evaporation and sunshine columns from the dataset permanently.

```python
df_1 = df.drop(['evaporation','sunshine'], axis = 1) # inplace = True
df_1
```

# 14- Find out the most rainy day for each city.

```python
most_rainy = df.loc[df.groupby('location')['rainfall'].idxmax().values][['location','date','rainfall']].set_index('location')
most_rainy
```

# 15- Filter out all the data for the city 'Albury' and then sort according to maxtemp column.

```python
df[df['location']=='Albury'].sort_values(by = 'maxtemp', ascending = False)
```

# 17- Remove the rows with NaN values in "windgustdir" column from the dataframe permanently.

```python
df.dropna(subset = ['windgustdir'],inplace = True)
df.reset_index(inplace = True)
df
```

# 18- Create a new dataframe, use "Location" column as the index of the dataframe, display the min, max, and median values of "evaporation" and "sunshine" columns in this dataframe.

```python
location = df.groupby('location').agg({'sunshine':['min','max','median'],'evaporation':['min','max','median']})
location
```
